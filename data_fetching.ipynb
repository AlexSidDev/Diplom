```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings

%matplotlib inline

warnings.filterwarnings('ignore')
```

```python
data_GOOG = yf.download('GOOG','1990-01-01','2019-08-01')
plt.plot(data_GOOG.index, data_GOOG['Adj Close'])
```

```python
USA_inflation = pd.read_excel("Inflation_rates.xlsx")
USA_inflation = USA_inflation.set_index('Year')
USA_inflation
```

```python
YEARS = USA_inflation.index.tolist()

def get_index_for_year(year: int):
    return YEARS.index(year)


def get_fixed_data(data: pd.DataFrame, inflations: pd.DataFrame):
    data_year = [row_index.date().year for row_index in data.index]
    
    data = data.reset_index()
    inflation_rates_for_year = { year : [rate for rate in inflations['Inflation_rate'][:get_index_for_year(year) + 1]]
                                for year in inflations['Year'].tolist()}
    
    for ind in range(len(data.index)):
        accumulated = data['Adj Close'][ind]
        
        for rate in inflation_rates_for_year[data_year[ind]][::-1]:
            accumulated /= rate
        data['Adj Close'][ind] = accumulated
    
    return data.set_index('Date')
```

```python
inflation_rates = (1 + USA_inflation['Inflation_rate']/100).reset_index()
data_GOOG_fixed = get_fixed_data(data_GOOG, inflation_rates[get_index_for_year(data_GOOG.index[0].date().year):])
fig, axes = plt.subplots(1, 2, figsize=(22, 6))
axes[0].plot(data_GOOG.index, data_GOOG['Adj Close'])
axes[1].set_ylim(axes[0].get_ybound())
axes[1].plot(data_GOOG.index, data_GOOG_fixed['Adj Close'])
axes[0].set_title("Google shares")
axes[1].set_title("Google shares in 2003's prices (without inflation influencing)")
```
