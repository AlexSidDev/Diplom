```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
data_GOOG = yf.download('GOOG','1990-01-01','2019-08-01')
plt.plot(data_GOOG.index, data_GOOG['Adj Close'])
```

```python
print(len(data_GOOG.index))
periods = 4
periods_len = int(len(data_GOOG.index)/periods)
print(periods_len)
fig, axes = plt.subplots(1, periods, figsize=(27, 7))
for period in range(periods):
    period_slice = slice(period * periods_len, (period + 1) * periods_len)
    data_GOOG_period = data_GOOG['Adj Close'][period_slice]
    data_ingex_period = data_GOOG.index[period_slice]
    axes[period].plot(data_ingex_period, data_GOOG_period)
    axes[period].xaxis.set_major_locator(ticker.MultipleLocator(400))
```

```python
data_IBM = yf.download('IBM','1990-01-01','2019-08-01')
plt.plot(data_IBM.index, data_IBM['Adj Close'])
```

```python
data_AAPL = yf.download('AAPL','1990-01-01','2019-08-01')
plt.plot(data_AAPL.index, data_AAPL['Adj Close'])
```

```python

# Create a subplot
periods = 4
periods_len = int(len(data_AAPL.index)/periods)
print(periods_len)
fig, axes = plt.subplots(2, periods, figsize=(22, 7))
for period in range(periods):
    period_slice = slice(period * periods_len, (period + 1) * periods_len)
    data_IBM_period = data_IBM['Adj Close'][period_slice]
    data_AAPL_period = data_AAPL['Adj Close'][period_slice]
    data_ingex_period = data_IBM.index[period_slice]
    axes[0][period].plot(data_ingex_period, data_IBM_period)
    axes[1][period].plot(data_ingex_period, data_AAPL_period)
```
